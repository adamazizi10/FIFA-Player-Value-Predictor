```python
df1 = pd.DataFrame(data_frame, columns=['Name', 'Wage', 'Value'])
def value_to_float(x):
    if type(x) == float or type(x) == int:
        return x
    if 'K' in x:
        if len(x) > 1:
            return float(x.replace('K', '')) * 1000
        return 1000.0
    if 'M' in x:
        if len(x) > 1:
            return float(x.replace('M', '')) * 1000000
        return 1000000.0
    if 'B' in x:
        return float(x.replace('B', '')) * 1000000000
    return 0.0
wage = df1['Wage'].replace('[\€,]', '', regex=True).apply(value_to_float)
value = df1['Value'].replace('[\€,]', '', regex=True).apply(value_to_float)
df1['Wage'] = wage
df1['Value'] = value
df1['difference'] = df1['Value'] - df1['Wage']
df1.sort_values('difference', ascending=False)
```

```python
import seaborn as sns
sns.set()

graph = sns.scatterplot(x='Wage', y='Value', data=df1)
graph
```

```python
from bokeh.plotting import figure, show
from bokeh.models import HoverTool

TOOLTIPS = HoverTool(tooltips=[
    ("index", "$index"),
    ("(Wage, Value)", "(@Wage, @Value)"),
    ("Name", "@Name")]
)
p = figure(title="Soccer 2019", x_axis_label='Wage', y_axis_label='Value', width=700, height=700, tools=[TOOLTIPS])
p.circle('Wage', 'Value', size=10, source=df1)
show(p)
```

```python
import pandas as pd

# Import the data using Pandas read_csv() function
data_frame = pd.read_csv('data.csv')

# Remove any irrelevant columns
data_frame = data_frame.drop(['Photo', 'Flag', 'Club Logo', 'Real Face', 'Jersey Number'], axis=1)

# Handle missing values
data_frame = data_frame.fillna(0)

# Remove duplicates
data_frame = data_frame.drop_duplicates()
```

```python
data_frame.describe()
```

```python
import seaborn as sns

# create a histogram of the 'Age' column
sns.histplot(data_frame['Age'])

# create a scatter plot of 'Value' vs. 'Wage'
sns.scatterplot(x='Wage', y='Value', data=data_frame)
```

```python
# create a correlation matrix of all numeric columns
data_frame_corr = data_frame.corr(numeric_only=True)

# create a heat map of the correlation matrix
sns.heatmap(data_frame_corr, cmap='coolwarm')
```
